#10. Compare Sales trend across Quarters using Order Date

#13. Compare Sales across Regions and also count number of orders.

```python
#13. Compare Sales across Regions and also count number of orders.
# The original error occurred because 'Order ID' was not included in the initial column selection.
# To calculate both sales and order count per region, we can use the .agg() function.
grped_region_stats = india_data.groupby("Region").agg(
    Total_Sales=('Sales', 'sum'),
    Number_of_Orders=('Order ID', 'count')
)

print(grped_region_stats)
```

```python
#14. Show average Profit per Product sold in each Region
avg_profit_per_product_region = india_data.groupby(["Region", "Product Name"])["Profit"].mean()
print(avg_profit_per_product_region)
```

```python
#15. Group orders by Region and Segment to analyze Shipping Costs.
shipping_costs_by_region_segment = india_data.groupby(["Region", "Segment"])["Shipping Cost"].sum()
shipping_costs_by_region_segment
```

```python
plt.pie(shipping_costs_by_region_segment,labels=shipping_costs_by_region_segment.index)
plt.title("Shiping cost by region and segment")
plt.show()
```

```python
#17. Compare Sales and Profit by Customer Name — top 10 customers only.
Top_10_customers = india_data.groupby("Customer Name").agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
)

print(Top_10_customers )
```

```python
#18. Show number of orders and average profit by Gender and Region.
orders_by_gender_region = india_data.groupby(["Gender", "Region"]).agg(
    Number_of_Orders=('Order ID', 'count'),
    Average_Profit=('Profit', 'mean')
)
orders_by_gender_region
```

```python
#19. Sales by Gender and Order Priority.
Total_sales=india_data.groupby(["Gender","Order Priority"])['Sales'].sum()
Total_sales
```

```python
#20. Group Customers by Manager or Team Lead and analyze total quantity ordered.
customer_manager_stats = india_data.groupby(["Customer Name","Manager"])["Quantity"].sum()
customer_manager_stats

```

```python
#21. Show total Quantity sold by Product Name and Sub-Category.
total_quantity_sold = india_data.groupby(["Product Name", "Sub-Category"])["Quantity"].sum()
total_quantity_sold
```

```python
#22. Analyze Sales by Product Category and Shipping Cost.
sales_by_category_shipping = india_data.groupby(["Product Name", "Category", "Shipping Cost"])["Sales"].sum()
sales_by_category_shipping
```

```python
#23. Which Sub-Category has the highest Profit-to-Sales ratio?
profit_to_sales_ratio = india_data.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).head(1)
profit_to_sales_ratio
```

#24. Group Products by Category and calculate average Sales and average Profit.

```python
#24. Group Products by Category and calculate average Sales and average Profit.
avg_sales_profit_by_category = india_data.groupby("Category")[["Sales", "Profit"]].mean()
avg_sales_profit_by_category
```

```python
#25. Show Sub-Categories that contributed over ₹1000 in total Profit.
sub_category_profit = india_data.groupby("Sub-Category")["Profit"].sum()
sub_category_profit=sub_category_profit[sub_category_profit>1000]
sub_category_profit
```

```python
plt.bar(sub_category_profit.index,sub_category_profit.values.flatten())
plt.xticks(rotation=90)
plt.title("Sub Category wise profit")
plt.grid(axis='y')
plt.show()
```
